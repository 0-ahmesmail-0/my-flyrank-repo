## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

# تعديل المسار ليرجع خطوتين فقط للخلف للوصول للـ Root
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

# 1. Total available content pool vs. pool showing a downward trend
total_pages = len(df)
declining_pages = len(df[df['trend_direction'] == 'down'])
declining_pct = (declining_pages / total_pages) * 100

# 2. High-exposure pages at risk (pages with >= 500 impressions in past 90 days that are declining)
high_exposure_declining = len(df[(df['trend_direction'] == 'down') & (df['impressions_90d'] >= 500)])

print(f"Total Pages Analysed: {total_pages:,}")
print(f"Declining Pages: {declining_pages:,} ({declining_pct:.1f}%)")
print(f"High-Exposure Declining Pages (>=500 impressions): {high_exposure_declining:,}")
```
